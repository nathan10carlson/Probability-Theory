```python
import matplotlib.pyplot as plt

# Number of intervals to show
num_intervals = 100

plt.figure(figsize=(8, 4))

for k in range(1, num_intervals + 1):
    left = 0
    right = 2 - 1/k
    plt.hlines(y=k, xmin=left, xmax=right, color='blue', linewidth=4)
    plt.text(right + 0.02, k, f'A_{k}', verticalalignment='center')

plt.xlabel('x')
plt.ylabel('Interval index k')
plt.title('Intervals A_k = [0, 2 - 1/k] and their growth')
plt.xlim(0, 2.1)
plt.ylim(0, num_intervals + 1)
plt.show()
```
